```python
# Visualizar mapa de calor
corr_inv = correlacion.iloc[::-1]
heatmap = plt.pcolor(correlacion, cmap = 'coolwarm', vmin = -1, vmax = 1)
plt.title('Mapa de calor')
plt.xticks(np.arange(0.5, long_col, 1), columnas)
plt.yticks(np.arange(0.5, long_col, 1), columnas[::-1])

for i in range(len(columnas)):
  for j in range(len(columnas)):
    val = round(corr_inv.iloc[i,j],2)
    plt.text(j + 0.5, i + 0.5, f'{val}',
             ha = 'center', va = 'center',
             color = 'white' if abs(val) > 0.6 else 'black')

   # print(val)

plt.colorbar(heatmap)
plt.tight_layout()
plt.show()
```

```python
# import from sklearn the linear regression model that will help us in this analysis.

from sklearn.linear_model import LinearRegression

# Define X (independent variable) and y (dependent variable) for linear regression
X = df_iris['sepal_length'].values.reshape(-1, 1) # Reshape to 2D array
y = df_iris['petal_length'].values

# create an empty linear regression model like below and give it a good variable name
linear_model = LinearRegression()

# to create the model, we use fit(X,y)
linear_model.fit(X, y)


y_pred = linear_model.predict(X)
plt.scatter(X, y, color = 'b', label='Actual data')
plt.plot(X, y_pred, color = 'r', label='Linear Regression')
plt.title('Linear Regression: Petal Length vs Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.legend()
plt.show()
```

```python
# The original cell contained R code, which is not executable in a Python kernel.
# This caused a SyntaxError. Below is the Python equivalent using matplotlib and scikit-learn.

# Ensure LinearRegression is imported, if not already in a prior cell's execution context.
from sklearn.linear_model import LinearRegression

# Based on the R code: plot(iris$petalos_long, iris$sepalos_long, ...)
# This suggests the x-axis is 'petal_length' and the y-axis is 'sepal_length'.

# Define X (independent variable) and y (dependent variable) for linear regression
X_r = df_iris['petal_length'].values.reshape(-1, 1) # x-axis: Petal Length
y_r = df_iris['sepal_length'].values             # y-axis: Sepal Length

# Create a new linear regression model for this specific axis orientation
linear_model_r = LinearRegression()
linear_model_r.fit(X_r, y_r)

# Predict the y values for the regression line
y_pred_r = linear_model_r.predict(X_r)

# Plotting the scatter plot and regression line using matplotlib
plt.figure(figsize=(7, 5)) # Equivalent to par(mfrow = c(1,1)) for a single plot and sets size
plt.scatter(X_r, y_r, color='#2ECC71', s=50, label='Actual data') # pch=19, col="#2ECC71", cex=1.2
plt.plot(X_r, y_pred_r, color='#2E86C1', linewidth=2.5, label='Linear Regression') # abline, col="#2E86C1", lwd=2.5

plt.title("Longitud de los PÉTALOS vs SÉPALOS (mm)") # main title
plt.xlabel("Longitud del Pétalo (mm)") # xlab
plt.ylabel("Longitud del Sépalo (mm)") # ylab

# Adding a subtitle if desired, often integrated into title or as text annotation in Python
# plt.text(X_r.min(), y_r.max() * 1.05, "Iris | Edgar Anderson´s Iris Data", fontsize=9, ha='left')

plt.legend()
plt.grid(True, linestyle='--', alpha=0.7) # Adding a grid for better readability
plt.show()
```
